```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
df.isnull().sum()
```

```python
## one hot encoding for categorical columns 
df_encode=pd.get_dummies(df,drop_first=True)
```

```python
df_encode
```

```python
## split the data into train and test 
##into independent(X) and dependent (y)
X=df_encode.drop(["price","bedrooms","furnishingstatus_semi-furnished"],axis=1)
y=df_encode["price"]
```

```python
X
```

```python
y
```

```python
## split the data into test and train
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=45)
```

```python
type(X)
```

```python
## INPUT FEATURES SHOULD ALWAYS BE A DATAFRAME 
```

```python
##TRAIN TE MODEL
```

```python
model=LinearRegression()
```

model

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
## predict the values from 
y_predict=model.predict(X_test)
```

```python
y_predict
```

```python
## evaluation of the model 
```

```python
mae=mean_absolute_error(y_test,y_predict
                       )
```

```python
mae
```

mse=mean_squared_error(y_test,y_predict)

```python
mse=mean_squared_error(y_test,y_predict)
```

```python
mse
```

```python
r2=r2_score(y_test,y_predict)
```

```python
r2
```

```python
import numpy as np
```

```python
mape=np.mean(np.abs((y_test-y_predict)/y_test))*100
```

```python
mape
```

```python
print("r2 score is :",r2)
print ("mape is :",mape)
print("mae is :",mae)
print("mse is :",mse)
```

```python
print("y intercept is ",model.intercept_)## c value 
print("slope is ",model.coef_[0])
```
